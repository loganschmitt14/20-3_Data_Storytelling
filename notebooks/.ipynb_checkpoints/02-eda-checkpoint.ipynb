# Drive for Show, Putt for Dough? #
### An analysis of PGA Tour data ###
Investigating the classic golf expression, which implies that long driving is all style and no substance, and that the real players earn their livings on the green. Answering the question: 
#### Is it more important to be a long-hitter with driver or great putter on the PGA Tour? 

## Preparations ##
Importing packages, setting styles, reading data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
palette = sns.color_palette()
sns.set(style = 'darkgrid', context = 'paper')
plt.style.use('dark_background')
sns.set_palette(palette)
plt.rcParams.update({'grid.linewidth' : 0.5, 'grid.alpha' : 0.75})
```

```python
pga_data = pd.read_csv('../data/interim/processed_data.csv', index_col = 0)
pga_data
```

## Analysis Plan ##
The data include strokes gained statistics for putting, tee-to-green, and total. Strokes gained effectively measures a player's efficiency relative to the field. Positive strokes gained in a category means the player is more efficient than average in that part of the game. 

The data also contain stats like average drive distance, driving accuracy as percentage of fairways hit, Greens-In-Regulation percentage, and scoring average measured in strokes per round. I'll investigate the relationship between putting stats and success (points, top 10's, wins, money) versus driving stats against the same metrics.

I'm also curious to know which correlates better: fairway percentage and greens-in-regulation (GIR) or driving distance and GIR. Conventional wisdom would dictate that being in the fairway leads to hitting more greens, but some might say that just being closer to the green, even in the rough, is better.

## Success Metrics ##
I have to decide how I'll evaluate success on Tour. I have data for money, wins, points, scoring average, and top-10 finishes.

```python
pga_data.nsmallest(5, ['scoring_avg'])
```

The data reflect the end of Tiger's prime, but he is still likely to appear near the top of a lot of lists. He averaged over 302 yards off the tee in 2007, when the tour average was 289 yards. His scoring average in 2007 is almost a full shot lower than anyone else on the list. 

```python
pga_data.nlargest(5, ['money'])
```

From 2007 through 2018, only 2 players had ever made more than $10 million in winnings during a single PGA Tour season. Tiger did it twice, both before 2010 (and a third time in 2005). I want to dive deeper into points, though. Tiger had well over 10 times as many points in 2007 as he did in 2009, despite making about the same amount of money. The Tour must have changed their approach to points.

```python
pga_data.loc[pga_data['year'] <= 2009].nlargest(5, ['points'])
```

Clearly the Tour changed the way points are awarded for the 2009 season. I should scale points and money by year so I can better control for changes like this. While I'm at it, I'll scale the other statistical columns so I can have a fully scaled DataFrame.

### Feature Centering ###
Feature centering by season will allow me to observe trends relative to the mean.

```python
from sklearn.preprocessing import StandardScaler

scaled_data = pga_data.copy()

scaler = StandardScaler(with_std = False)
years = scaled_data['year'].unique()
scale_cols = ['money', 'points', 'scoring_avg', 'avg_drive', 'fwy%', 'gir%', 'sg_putt', 'sg_t2g', 'sg_tot']

import warnings
warnings.filterwarnings('ignore')

for year in years:
    year_data = scaled_data[scaled_data['year'] == year]
    year_data[scale_cols] = scaler.fit_transform(year_data[scale_cols])
    scaled_data[scaled_data['year'] == year] = year_data
    
scaled_data
```

I want to examine how the different metrics for success correlate with one another. I'm curious to see how money and points relate to one another and to rounds played. I need to choose how I'll measure "success."

```python
success_cols = ['rounds', 'scoring_avg', 'top_10', 'money', 'points']

correlation_matrix = scaled_data[success_cols].corr()

column_labels = ['Rounds', 'Scoring Average', 'Top 10s', 'Money', 'Points']
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, xticklabels=column_labels, yticklabels = column_labels)
plt.title('Correlation Heatmap Between Success Metrics on Tour')
plt.show()
```

Rounds correlates weakly with the rest of the statistics. I suspected this might be the case. Many of the best players actually play fewer events because they earn enough points to make the playoffs by winning and earning top-10 finishes. The golfers who are less likely to win any given event are more likely to try and amass points toward playoff status by playing more events, even if they tend to finish lower on the leaderboard.

It is no surprise that golfers who score lower earn more top-10 finishes, and therefore make more money and earn more points.

#### Conclusion ####
It seems fair to say that players with a lower scoring average will be more successful on Tour, so I'll examine whether putting or driving has a greater effect on scoring average. This makes the result a bit more applicable to the average golfer as well, who isn't competing for money.

## Driving Trends ##
I want to check out driving trends next. How far are Tour players hitting the ball on average? Are they hitting fairways? Have those numbers changed over time?

### Driving Stat Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.subplots_adjust(top=0.8)

sns.histplot(data = pga_data, x = 'avg_drive', binwidth = 5, ax = axes[0])
axes[0].set_title('Histogram of Driving Distance')
axes[0].set_xlabel('Average Driving Distance (Yards)')
axes[0].grid(axis = 'x')

sns.histplot(data = pga_data, x = 'fwy%', bins = 12, ax = axes[1])
axes[1].set_title('Histogram of Driving Accuracy')
axes[1].set_xlabel('Driving Accuracy (% of Fairways Hit)')
axes[1].grid(axis = 'x')
plt.show()
```

The histograms for driving distance and driving accuracy are remarkably similar. Most players seem to hit the ball around 290 yards and find just under 2/3 fairways. Has this always been the case or have these numbers changed with time? People always talk about the distance problem in golf. Had that begun by 2018?

### Driving Stat Trends Over Time

```python
# Calculating yearly averages for driving distance, driving accuracy, and greens-in-regulation
avg_drive_yearly = (pga_data
                    .groupby('year')
                    .avg_drive
                    .mean()
                    .round(0)
                    .reset_index(drop = False))

fwy_pct_yearly = (pga_data
                  .groupby('year')['fwy%']
                  .mean()
                  .round(0)
                  .reset_index(drop = False))

gir_pct_yearly = (pga_data
                  .groupby('year')['gir%']
                  .mean()
                  .round(0)
                  .reset_index(drop = False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
fig.subplots_adjust(top=0.8)


start_yards = (avg_drive_yearly['year'].iloc[0], avg_drive_yearly['avg_drive'].iloc[0])
end_yards = (avg_drive_yearly['year'].iloc[-1], avg_drive_yearly['avg_drive'].iloc[-1])


start_pct = (fwy_pct_yearly['year'].iloc[0], fwy_pct_yearly['fwy%'].iloc[0])
end_pct = (fwy_pct_yearly['year'].iloc[-1], fwy_pct_yearly['fwy%'].iloc[-1])

sns.lineplot(data = avg_drive_yearly, x = 'year', y = 'avg_drive', marker = '.', linewidth = '2', ax = axes[0])
axes[0].set_title('Average PGA Tour Driving Distance Over Time')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Average Driving Distance (Yards)')
axes[0].set_ylim(275, 300)
axes[0].annotate(f'{start_yards[1]:.0f} yards', start_yards, textcoords="offset points", xytext=(15,15), ha='center')
axes[0].annotate(f'{end_yards[1]:.0f} yards', end_yards, textcoords="offset points", xytext=(-15,5), ha='center')


sns.lineplot(data = fwy_pct_yearly, x = 'year', y = 'fwy%', marker = '.', linewidth = '2', ax = axes[1])
axes[1].set_title('Average PGA Tour Driving Accuracy Over Time')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Average Driving Accuracy (%)')
axes[1].set_ylim(50, 70)
axes[1].annotate(f'{start_pct[1]:.1f}%', start_pct, textcoords="offset points", xytext=(10,10), ha='center')
axes[1].annotate(f'{end_pct[1]:.1f}%', end_pct, textcoords="offset points", xytext=(-5,10), ha='center')

plt.show()
```

The distance and accuracy trends are generally opposite to one another. Players hit the ball about 3% farther on average in 2018 compared to 2010 but they hit the fairway about 2% less often. I will investigate if trading accuracy for distance pays off. My suspicion is that sacrificing accuracy must be worth it, or players wouldn't be pursuing distance at its expense. After all, they and their coaches have these stats too.

### Distance or Accuracy?
Which matters more for strokes gained: tee-to-green? 

In this case, I'm examining whether driving accuracy or driving distance correlates more strongly with a player efficiently getting the ball to the green.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax = sns.scatterplot(data = scaled_data, x = 'avg_drive', y = 'fwy%', size = 'sg_t2g', hue = 'sg_t2g', alpha = .75, palette = sns.color_palette('icefire', as_cmap=True), sizes=(10, 100))
ax.set_title('Scatterplot of Driving Accuracy vs Driving Distance Correlated with Strokes Gained: Tee-to-Green')
ax.set_xlabel('Driving Distance Relative to Yearly Average')
ax.set_ylabel('Driving Accuracy Relative to Yearly Average')
ax.vlines(x = 0, ymin = -30, ymax = 30, color = 'w', linewidth = 2)
ax.hlines(y = 0, xmin = -30, xmax = 30, color = 'w', linewidth = 2)
ax.set_ylim(-30, 30)
ax.set_xlim(-30, 30)
ax.legend_.set_title('SG: T2G')
plt.show()
```

There is a huge cluster around the mean, but it's pretty clear that the higher strokes-gained players are along the upper-right edge of the cluster, meaning they are above average in one category or the other. The distribution seems to be favoring the right side of the graph, therefore leaning toward longer hitters. Being below average in either category corresponds well with a lower strokes gained. Interestingly, no player was able to hit the ball 20 yards farther than the average player while also being above average in fairway percentage. On the flip side, no player had above-average length off the tee and exceeded the average fairway percentage by 10%. It is really hard to get the best of both worlds.

Overall, this scatterplot doesn't decisively show whether accuracy or distance has the edge in terms of strokes gained.

### Statistical Correlations

I want to see how a host of driving stats correlate with measures of success, keeping in mind that scoring average is my metric of interest.

```python
driving_cols = ['avg_drive', 'fwy%', 'gir%', 'sg_t2g', 'scoring_avg']


correlation_matrix = scaled_data[driving_cols].corr()

column_labels = ['Average Drive\n Distance', 'Fairway %', 'Greens in\n Regulation %', 'Strokes Gained:\n Tee-to-Green', 'Scoring\n Average', ]
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, xticklabels=column_labels, yticklabels = column_labels)
plt.title('Correlation Heatmap Between Driving Statistics and Scoring Average on Tour')
plt.show()
```

There are a number of correlations to point out here. 
* As observed previously, driving distance correlates negatively with fairway percentage.
* Strokes Gained: Tee to Green correlates more strongly with driving distance than driving accuracy.
* Drive distance and fairway percentage share a very similar correlation with Greens in Regulation, which was shown in the scatterplot above as well.
* Scoring average correlates more strongly with driving distance than driving accuracy.
* The correlation between strokes gained: tee-to-green and greens in regulation is lower than I'd expect.
    * I imagine this may have to do with proximity to the hole being factored into strokes gained

## Putting Trends ##
I don't expect putting numbers to really change over time like driving numbers have.

![A professional golfer attempts a putt. Photo by Richard Stott on Unsplash](../images/putt1.jpg)

```python
avg_sg_putt_yearly = (pga_data                   
                        .groupby('year')
                        .sg_putt
                        .mean()
                        .round(3)
                        .reset_index(drop = False))
```

### Strokes Gained: Putting Distribution

```python
fig, axes = plt.subplots()
sns.histplot(data = pga_data, x = 'sg_putt')
axes.set_title('Histogram of Strokes Gained: Putting')
axes.set_xlabel('Strokes Gained: Putting')
axes.grid(axis = 'x')
```

There's a negative skew to the distribution, which does make sense. Most Tour pros are pretty good putters already. It's actually impossible to be that much better than a tour pro in this stat, since you would still need to hit at least one putt per hole (hole-outs don't affect this stat). It is much more feasible to be a worse putter and make up for it in other ways.

### Average Strokes Gained: Putting Over Time
Strokes Gained stats are computed as performance relative to the field. In light of that fact, I'm expecting to see a horizontal line at 0 when looking at average strokes gained.

```python
start_sg = (avg_sg_putt_yearly['year'].iloc[0], avg_sg_putt_yearly['sg_putt'].iloc[0])
end_sg = (avg_sg_putt_yearly['year'].iloc[-1], avg_sg_putt_yearly['sg_putt'].iloc[-1])

fig, axes = plt.subplots()
sns.lineplot(data = avg_sg_putt_yearly, x = 'year', y = 'sg_putt', marker = '.', linewidth = '2')
axes.set_title('Average PGA Tour Strokes Gained: Putting Over Time')
axes.set_xlabel('Year')
axes.set_ylabel('Average Strokes Gained: Putting')
axes.set_ylim(-.25, .25)
axes.annotate(f'{start_sg[1]:.3f}', start_sg, textcoords="offset points", xytext=(10,5), ha='center')
axes.annotate(f'{end_sg[1]:.3f}', end_sg, textcoords="offset points", xytext=(-15,5), ha='center')
plt.show()
```

As expected, this is pretty much a flat line. It does stay slighly above 0, which probably means that some of the people being factored into the "field" statistic aren't full-time Tour players and are thus absent from this dataset. It makes sense that the average Tour player is slightly gaining strokes relative to a field that includes a couple amateurs and sponsor's exemptions.

### Statistical Correlations
It's only fair to see how strokes gained: putting correlates with success on tour.

```python
putting_cols = ['sg_putt', 'scoring_avg']

correlation_matrix = scaled_data[putting_cols].corr()

column_labels = ['Strokes Gained: Putting', 'Scoring\n Average']
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, xticklabels=column_labels, yticklabels = column_labels)
plt.title('Correlation Heatmap Between Strokes Gained: Putting\n and Scoring Average on Tour')
plt.show()
```

Strokes Gained: Putting doesn't correlate as strongly with scoring average as I expected. I checked against points, money, and top 10's and the correlation was even weaker.

## Putting vs Driving
Time for the head-to-head comparison. Which is more important for scoring -- being a good putter or a bomber? If I had Strokes Gained: Off-The-Tee, I'd compare it to Strokes Gained: Putting. Sadly, that statistic wasn't available for all the years. I'll use driving distance instead. Strokes Gained: Tee-To-Green can't be used here because it includes approach shots.

```python
lowest_scoring_players = (scaled_data
                          .loc[scaled_data['scoring_avg'] < scaled_data['scoring_avg'].quantile(0.25)]
                          .set_index(['name', 'year'])
                          .assign(score_status = 0))

best_drivers = (scaled_data
                .loc[scaled_data['avg_drive'] > scaled_data['avg_drive'].quantile(0.75)]
                .set_index(['name', 'year'])
                .assign(drive_status = 2)
                .loc[:,['drive_status']]
               )

best_putters = (scaled_data
                .loc[scaled_data['sg_putt'] > scaled_data['sg_putt'].quantile(0.75)]
                .set_index(['name', 'year'])
                .assign(putt_status = 1)
                .loc[:,['putt_status']]
               )

best_scorers = (lowest_scoring_players
                .join([best_putters, best_drivers], how = 'left')
                .fillna(0)
                .assign(status = lambda x: x['score_status'] + x['putt_status'] + x['drive_status'])
                .assign(status_category=lambda x: pd.cut(x['status'], bins=4, labels=['Top 25% Scoring Only', 'Top 25% Putters', 'Top 25% Drivers', 'Top 25% All Categories']))
               )
```

```python
# Create a plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.set_ylim(-1.5, 1.5)
ax.set_xlim(-30, 30)

ax.vlines(x = 0, ymin = -4, ymax = 4, color = 'w', linewidth = 2, label='_nolegend_')
ax.hlines(y = 0, xmin = -40, xmax = 40, color = 'w', linewidth = 2, label='_nolegend_')

ax.set_title('Driving Stats vs. Putting Stats Among Top Quartile of Scorers')
ax.set_xlabel('Driving Distance Above Average')
ax.set_ylabel('Strokes Gained: Putting Above Average')

# Add data
sns.scatterplot(x='avg_drive', y='sg_putt', data=best_scorers, s = 40, alpha = 0.85, hue = 'status_category', linewidth = 0, palette = sns.color_palette('Set2', 4), legend = 'full')

# Creating annotations for specific points
best_putter_raw = pga_data.loc[scaled_data['sg_putt'] > 1].reset_index(drop=True)
best_putter_scaled = scaled_data.loc[scaled_data['sg_putt'] > 1].reset_index(drop = True)
putt_coords = (best_putter_scaled['avg_drive'].values[0], best_putter_scaled['sg_putt'].values[0])

worst_driver_raw = pga_data.loc[scaled_data['avg_drive'] < -25].reset_index(drop=True)
worst_driver_scaled = scaled_data.loc[scaled_data['avg_drive'] < -25].reset_index(drop = True)
driver_coords = (worst_driver_scaled['avg_drive'].values[0], worst_driver_scaled['sg_putt'].values[0])

ax.annotate(f"{best_putter_raw['name'].values[0]}\n SG: Putting {best_putter_raw['sg_putt'].values[0]}\n Avg Drive: {best_putter_raw['avg_drive'].values[0]:.0f}\n Avg Score: {best_putter_raw['scoring_avg'].values[0]:.1f}", putt_coords, textcoords="offset points", xytext=(86,4), ha='center', arrowprops=dict(facecolor = 'g', edgecolor = 'k'))
ax.annotate(f"{worst_driver_raw['name'].values[0]}\n SG: Putting {worst_driver_raw['sg_putt'].values[0]:0.2f}\n Avg Drive: {worst_driver_raw['avg_drive'].values[0]:.0f}\n Avg Score: {worst_driver_raw['scoring_avg'].values[0]:.1f}", driver_coords, textcoords="offset points", xytext=(82, 15), ha='center', arrowprops=dict(facecolor = 'g', edgecolor = 'k'))

# Create legend 
handles, labels = ax.get_legend_handles_labels()
category_counts = best_scorers['status_category'].value_counts()
labels=[f'{category} ({count})' for category, count in category_counts.items()]

ax.legend(handles, labels, loc = 'lower left', title='Category')


plt.show()
```

This graph is fascinating because it just features the 25% of players with the lowest scoring average relative to other players in a given year. It is quite impressive that a handful of these players are losing strokes putting and hitting shorter drives but are still scoring so well. These are the players in the lower-left quadrant.

Shown in pink are the players who are also in the upper quartiles for driving distance and Strokes Gained: Putting. There are just 60 players in this group out of over 2,000 in the dataset. This group includes Jason Day's 2016 season, which does stand out in a very impressive way. He was the only player to gain more than one stroke per round above average on the field with his putting. He did so while driving the ball about 13 yards farther than the average player too, which puts him near the 90th percentile in driving distance.

The players in orange aren't long hitters but are in the top 25% of putters. Corey Pavin seems to be making a statement about his own opinion on the drive vs putt debate. He was 25 yards shorter off the tee than the average tour player, but picked up nearly a stroke per round on the field putting. He embodies "putt for dough" here.

The blue dots refer to the bombers who aren't excellent putters. There aren't really any noteworthy outliers like these two in terms of driving distance. Nobody managed to consistently hit the ball really far while being a terrible putter.

The green dots represent players who are in the top quartile for scoring average without being in the top quartile for either putting or driving distance. This is the largest cluster

As far as settling the putting vs driving debate, this graph doesn't show a clear trend, which is an answer in and of itself. The clusters are roughly equivalent in each of the three "positive" quadrants, where a player would be above average in at least one of the metrics. Unsurprisingly, most of the best players seem to be above average in driving distance and strokes gained: putting. 


## Conclusion ##

There are a couple takeaways here. The first is that clearly, the adage of "drive for show, putt for dough" isn't true for everyone. There are plenty of great players who are below average putters but still perform really well on Tour. On the flip side, the expression holds true for a lot of shorter hitters. There are guys like Corey Pavin who earn their living on Tour by keeping the ball in play and cleaning up around the greens. The reality is that while only a handful of guys excel in both categories, it is really important not to fall below average in both. Iron play and short game can only take you so far on Tour. The game is getting longer, and those who aren't carving out a niche for themselves as outstanding putters should really consider adding some length if they want to remain competetive.
